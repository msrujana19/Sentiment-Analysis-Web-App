```python
import pandas as pd
from transformers import pipeline
```

```python
reviews=df['review_content'].to_list()
reviews[:5]
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
sentiment_pipeline("This product is great!")

```

```python
# Apply sentiment analysis to all reviews
df['sentiment'] = df['review_content'].apply(lambda x: sentiment_pipeline(x)[0]['label'])
```

```python
def safe_sentiment(review):
    try:
        # Truncate to 512 characters (approximate safe token count)
        return sentiment_pipeline(review[:512])[0]['label']
    except:
        return "ERROR"

df['sentiment'] = df['review_content'].apply(safe_sentiment)
```

```python
import re
from wordcloud import WordCloud
from nltk.corpus import stopwords
import nltk

# Download stopwords if not already present
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation/numbers
    words = text.split()
    words = [word for word in words if word not in stop_words]  # remove stopwords
    return ' '.join(words)
```

```python
df['cleaned_review'] = df['review_content'].apply(clean_text)
df['cleaned_review'].head()
```

```python
# Join all cleaned review texts by sentiment
positive_text = ' '.join(df[df['sentiment'] == 'POSITIVE']['cleaned_review'])
negative_text = ' '.join(df[df['sentiment'] == 'NEGATIVE']['cleaned_review'])

# Create and display Word Clouds
wordcloud_pos = WordCloud(width=800, height=400, background_color='white').generate(positive_text)
wordcloud_neg = WordCloud(width=800, height=400, background_color='black', colormap='Reds').generate(negative_text)

# Plot Positive
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_pos, interpolation='bilinear')
plt.title('Word Cloud - Positive Reviews')
plt.axis('off')
plt.tight_layout()
plt.show()

# Plot Negative
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_neg, interpolation='bilinear')
plt.title('Word Cloud - Negative Reviews')
plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import torch
import seaborn as sns
import matplotlib.pyplot as plt
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Step 1: Load pre-trained model and tokenizer
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, output_attentions=True)
model.eval()
```

```python
def get_attention_weights(text):
    # Tokenize input
    inputs = tokenizer(text, return_tensors="pt", truncation=True)
    
    # Run model and get attention outputs
    with torch.no_grad():
        outputs = model(**inputs)
        attentions = outputs.attentions  # List of attention tensors from each layer

    # Use last layer's attention, average over heads
    last_layer_attention = attentions[-1]               # Shape: [1, heads, tokens, tokens]
    mean_attention = last_layer_attention.mean(dim=1)   # Average over heads ‚Üí [1, tokens, tokens]
    
    tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
    return tokens, mean_attention[0].cpu()  # Remove batch dimension and move to CPU
```

```python
def plot_attention(tokens, attention):
    plt.figure(figsize=(10, 8))
    sns.heatmap(attention.numpy(), 
                xticklabels=tokens, 
                yticklabels=tokens, 
                cmap="viridis", 
                square=True, 
                linewidths=0.5)
    plt.title("Attention Heatmap (Last Layer, Averaged Heads)")
    plt.xlabel("Token")
    plt.ylabel("Token")
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
sample_text = "I absolutely love this product! It's amazing and works like a charm."
tokens, attention = get_attention_weights(sample_text)
plot_attention(tokens, attention)
```
